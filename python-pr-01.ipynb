```python
import re
```

```python
file = open('chat.txt', encoding='utf-8')
file.readline()
newfile = open('cleandata.txt','w', encoding='utf-8')
raw = []
count=0
for index, line in enumerate(file.readlines()):
    x = re.search(r'(.+), (\d+:\d+) (.*) - (.*): (.*)',line)

    if x == None:
        if len(raw) < 0:
            count+=1
            raw[index-count][-1]+= line.strip()
    else:
        raw.append([*x.groups()])
for line in raw:
    newfile.write(','.join(line)+'\n')

file.close()
newfile.close()
```
